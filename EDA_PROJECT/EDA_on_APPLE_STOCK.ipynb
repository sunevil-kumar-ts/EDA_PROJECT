```python
# 12. Trend of Closing Price over Time
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

plt.figure(figsize=(10,5))
sns.lineplot(x=data1['Date'], y=data1['Close'])

plt.title("Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")

plt.show()
```

The closing price shows an overall **upward trend with minor fluctuations**, indicating steady long-term growth.


```python
#13. Open Price vs Close Price
# -----------------------------
plt.figure(figsize=(10,5))
plt.plot(data1['Date'], data1['Open'], label='Open')
plt.plot(data1['Date'], data1['Close'], label='Close')
plt.title("Open vs Close Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

The **Open and Close prices move closely together**, indicating consistent daily trading behavior with only small price changes during the day.


```python
#14. Distribution of Closing Price
# -----------------------------
plt.figure(figsize=(8,5))
sns.histplot(data1['Close'], bins=30, kde=True)
plt.title("Distribution of Closing Price")
plt.show()
```

The closing prices are distributed across a range with slight skewness, indicating most values fall within a specific price band with some variation.

```python
# 15. Outliers in Closing Price
# -----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x=data1['Close'])
plt.title("Boxplot of Closing Price")
plt.show()
```

The boxplot shows **few outliers in the closing price**, indicating occasional extreme values but overall stable price distribution.


```python
# 16 . Relationship between Open and Close
# -----------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x=data1['Open'], y=data1['Close'])
plt.title("Open vs Close")
plt.show()
```

The scatter plot shows a **strong positive relationship between Open and Close prices**, indicating that both move closely together.


```python
# 17. Relationship between High and Low
# -----------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x=data1['High'], y=data1['Low'])
plt.title("High vs Low")
plt.show()

```

The scatter plot shows a **strong positive correlation between High and Low prices**, indicating consistent daily price ranges.


```python
# 18. Trading Volume Over Time
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.lineplot(x=data1['Date'], y=data1['Volume'])

plt.title("Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.show()
```

The trading volume shows **significant fluctuations over time**, indicating varying levels of market activity and investor interest.


```python
# 19. Distribution of Volume
plt.figure(figsize=(8,5))
sns.histplot(data1['Volume'], bins=30, kde=True)
plt.title("Distribution of Trading Volume")
plt.show()
```

The trading volume is **unevenly distributed with noticeable skewness**, indicating periods of both low and high market activity.


```python
#20. Outliers in Volume

plt.figure(figsize=(8,5))
sns.boxplot(x=data1['Volume'])
plt.title("Boxplot of Volume")
plt.show()
```

The boxplot shows **presence of outliers in trading volume**, indicating occasional spikes in market activity.


```python
# 21. Stock Price Volatility
# -----------------------------
data1['Volatility'] = data1['High'] - data1['Low']

plt.figure(figsize=(10,5))
plt.plot(data1['Date'], data1['Volatility'])
plt.title("Stock Price Volatility")
plt.xlabel("Date")
plt.ylabel("High - Low")
plt.show()

```

The volatility shows **fluctuating patterns over time**, indicating periods of both stable and highly variable stock prices.
